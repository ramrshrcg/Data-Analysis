```python
from sklearn.datasets import make_classification
```

```python
import pandas as pd

x,y=make_classification(n_samples=1000, n_features=2, n_clusters_per_class=1, weights=[.90],
                    random_state=12,n_redundant=0,)


df1= pd.DataFrame(x,columns=['c1','c2'])
df2= pd.DataFrame(y,columns=['target'])
df= pd.concat([df1, df2], axis=1)
df.to_csv('imbalanced_data.csv', index=False)


```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['c1'], df['c2'], c=df['target'])
plt.show()
```
