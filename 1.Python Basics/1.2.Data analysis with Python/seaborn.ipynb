```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x= 'total_bill', y='tip', data=tips,hue='smoker', style='time',palette='deep',)
plt.show()
```

```python
sns.lineplot(x='size', y='total_bill', data=tips, estimator='mean')
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tips, )
plt.show()
```

```python
 #boxplot
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
plt.show()
```

```python
# violin plot
sns.violinplot(x='day', y='total_bill', data=tips,)
plt.show()
```

```python
#histogram
sns.histplot(tips['total_bill'], kde=True, bins=30)
sns.histplot(data=tips, x='total_bill',kde=True)
plt.show() 
```

```python
#kde plot
sns.kdeplot(tips['total_bill'], fill=True)
  
plt.show()
```

```python
#pairplot
sns.pairplot(tips, diag_kind='kde')
plt.show()
```

```python
#heatmap
corr=tips[['total_bill','tip','size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
